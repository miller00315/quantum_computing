```python
# importação das classes que serão instanciadas como objeto
from Neuraline.QuantumPhysics.quantum_computing import QUANTUM_COMPUTER # classe de acesso ao computador quântico
from Neuraline.QuantumPhysics.quantum_computing import QuantumCircuit # classe para construção do circuito quântico
from Neuraline.Utilities.data import ConvertData # classe para leitura e conversão de dados
from Neuraline.Utilities.chart import Chart # classe para plotagem de dados gráficos
convert_data, chart = ConvertData(), Chart() # instanciação dos objetos das classes
```

## Construção do Circuito Quântico

```python
quantum_circuit = QuantumCircuit( # instanciação do objeto do circuito
    architecture=QUANTUM_COMPUTER # arquitetura de acesso remoto ao computador quântico real
)
quantum_circuit.setQubits(nqubits=2) # seta 2 qubits (bits quânticos) na construção do circuito
# código para a geração da lista com os endereços dos arquivos da primeira classe
url_paths = convert_data.getFileList(dir_path='./data/images/cat_and_dog/training/gato/') # captura dos arquivos no diretório
for path in url_paths: # loop para adicionar os arquivos da primeira classe um a um no circuito
    quantum_circuit.addFileData(data=path) # atribuição dos endereços das amostras de entrada
    quantum_circuit.addOutputData(data='gato') # atribuição dos rótulos de saída para cada entrada
# código para a geração da lista com os endereços dos arquivos da segunda classe
url_paths = convert_data.getFileList(dir_path='./data/images/cat_and_dog/training/cachorro/') # captura dos arquivos no diretório
for path in url_paths: # loop para adicionar os arquivos da segunda classe um a um no circuito
    quantum_circuit.addFileData(data=path) # atribuição dos endereços das amostras de entrada
    quantum_circuit.addOutputData(data='cachorro') # atribuição dos rótulos de saída para cada entrada
# código de inserção dos barramentos quânticos
result1 = quantum_circuit.addEncoderBarrierX() # barramento para conversão de dados clássicos em quânticos
result2 = quantum_circuit.addDecoderBarrierY() # barramento para conversão de dados quânticos em clássicos
if result1 and result2: print('Circuito quântico construído com sucesso.') # mensagem de sucesso
else: print('Erro na construção do circuito quântico.') # mensagem de fracasso
```

## Exibição do Circuito Quântico Montado

```python
result = quantum_circuit.showCircuit() # exibição do circuito quântico
if result: print('Engenharia quântica gravada com sucesso.') # mensagem de sucesso
else: print('Erro na gravação da engenharia quântica.') # mensagem de fracasso
```

## Medição e Conversão do Resultado
### Teste da Primeira Classe (primeiro estado quântico convertido em clássico)

```python
# código para a classificação da imagem de entrada em um dos estados quânticos possíveis
url_path = './data/images/cat_and_dog/test/gato/image01.jpg' # endereço da imagem a ser classificada
result = quantum_circuit.measurement( # medição do estado quântico
    repetitions=1024, # número referente a quantidade de medidas
    data=url_path # atribuição do endereço de entrada da medição
)
# código para capturar as chaves e valores do dicionário de resultado
x, y = list(result.keys()), list(result.values()) # x: estado quântico, y: valor probabilístico
print('Imagem a ser classificada:') # mensagem descritiva
# código para plotagem gráfica da imagem a ser classificada
chart.plotIMAGE( # plotagem da imagem que será classificada
    image=url_path # atribuição do endereço da imagem a ser classificada
)
print('Resultado probabilístico:') # mensagem descritiva
chart.plotBAR( # plotagem do gráfico (de barras) probabilístico
    x=x, # valores do eixo x (horizontal)
    y=y, # valores do eixo y (vertical)
    x_label='Estado Quântico', # descrição do eixo x (horizontal)
    y_label='Probabilidade %', # descrição do eixo y (vertical)
    title='Classificação Entre Gato e Cachorro', # descrição/título do gráfico
    bar_values=True # habilita a exibição dos valores nas barras
)
# mensagem descritiva com a classificação quântica convertida para clássica
print(f'Resultado classificativo do estado clássico: {quantum_circuit.getResult().upper()}')
```
